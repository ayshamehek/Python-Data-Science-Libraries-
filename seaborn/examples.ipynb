Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Scatter Plot")
plt.show()
```

Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

sns.lineplot(x="year", y="passengers", data=flights)

plt.title("Passengers Over Years")
plt.show()
```

Basic Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

sns.lineplot(x="year", y="passengers", data=flights)

plt.title("Passengers by Year")
plt.show()
```

Line Plot with Different Colors (Hue)

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

sns.lineplot(x="year", y="passengers", hue="month", data=flights)

plt.title("Passengers by Month")
plt.show()
```

Student Marks Trend

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    "Exam": [1, 2, 3, 4, 5],
    "Marks": [65, 70, 75, 80, 90]
})

sns.lineplot(x="Exam", y="Marks", data=data)

plt.title("Student Marks")
plt.show()
```
