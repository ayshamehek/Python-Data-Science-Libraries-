creating data from dictionary

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "London", "Paris"]
}

df = pd.DataFrame(data)

print(df)
```

Find Average Salary

```python
import pandas as pd

data = {
    "Employee": ["A", "B", "C", "D"],
    "Salary": [25000, 30000, 40000, 35000]
}

df = pd.DataFrame(data)

average = df["Salary"].mean()

print("Average Salary =", average)
```

Merge Two DataFrames

```python
import pandas as pd

students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"]
})

marks = pd.DataFrame({
    "ID": [1, 2, 3],
    "Marks": [90, 85, 95]
})

result = pd.merge(students, marks, on="ID")

print(result)
```
